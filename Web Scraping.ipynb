```python
import requests
from bs4 import BeautifulSoup

URL = 'https://priyoshop.com/grocery-items'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
productNames = soup.find_all('h2', class_='product-title')
```

```python
for productName in productNames:
    print(productName, end='\n')
```

```python
productPrices = soup.find_all('div', class_='prices')
```

```python
for productPrice in productPrices:
    print(productPrice, end='\n')
```

```python
import xlsxwriter
workbook = xlsxwriter.Workbook('dataset.xlsx')
worksheet = workbook.add_worksheet()
row = 1
column = 0
worksheet.write(row-1, column, "productName")
for productName in productNames:
    productTitle = productName.find('a')
    worksheet.write(row, column, productTitle.text)
    print(productTitle.text)
    row += 1

row=1
column = 1
worksheet.write(row-1, column, "productPrice")
for productPrice in productPrices:
    productCost = productPrice.find('span', class_='price actual-price')
    worksheet.write(row, column, productCost.text[3:])
    print(productCost.text[3:])
    row += 1
workbook.close()
```
