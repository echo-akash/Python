```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Create Graph
G = nx.Graph() #G is graph of no nodes and edges
```

```python
#Add Nodes
G.add_node(1) #Add 1 node
G.add_nodes_from([2,3]) #add list of nodes
G.nodes #View nodes
```

```python
#Add Edge
G.add_edge(1,2) #Add edge one by one
G.add_edges_from([(3,4),(5,6)]) #Add edge from list
G.edges #View edges

list(G.adj[1])  # list(G.neighbors(1))
G.degree[1]  # the number of edges incident to 1
```

```python
#remove node or edge
G.remove_node(2)
G.remove_edge(2,3)
G.clear() #remove everything
```

```python
#Access edge and neighbour
G.edges[1, 2]['color'] = "red"
G.add_weighted_edges_from([(1, 2, 0.125), (1, 3, 0.75), (2, 4, 1.2), (3, 4, 0.375)])

for n, nbrs in G.adj.items():
    for nbr, eattr in nbrs.items():
        wt = eattr['weight']
        if wt < 0.5: print('(%d, %d, %.3f)' % (n, nbr, wt))
            
for (u, v, wt) in G.edges.data('weight'):
    if wt < 0.5: print('(%d, %d, %.3f)' % (u, v, wt))
```
